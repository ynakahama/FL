```python
def two_plot_FL_Loss(global_performance1,global_performance2):
    print("=== Global and Local (Client 1) Test Loss ===")
    n_rounds = len(global_performance1)
    plt.plot(range(1, n_rounds+1), global_performance1, label="Random")
    plt.plot(range(1, n_rounds+1), global_performance2, label="FedCS")
    plt.xlabel('Rounds')
    plt.ylabel('Test loss')
    plt.legend()
    plt.show()
def two_plot_FL_Accuracy(global_performance1,global_performance2):
    n_rounds = len(global_performance1)
    print("=== Global and Local (Client 1) Test Accuracy ===")
    plt.plot(range(1, n_rounds+1), global_performance1, label="Random")
    plt.plot(range(1, n_rounds+1), global_performance2, label="FedCS")
    plt.xlabel('Rounds')
    plt.ylabel('Test accuracy')
    plt.legend()
    plt.show()
```
